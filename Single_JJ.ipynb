```python
import numpy as np
from matplotlib import pyplot as plt

from Objects import JJ, Ib
from FunctionCompiler import *
```

```python
JJ1 = JJ.JJ(loc=[1, 0])
ib = Ib.Ib(loc=[1])
```

```python
fc = FunctionCompiler([JJ1, ib], 'Ib1.val', 4)
fc.compile()
```

```python
t = np.arange(0, 20, 0.1)
sol = fc.solve(t, np.zeros(2))
```

```python
P_JJ1 = JJ1.get_data("P", t, sol)
V_JJ1 = JJ1.get_data("V", t, sol)

plt.plot(P_JJ1[:,0], P_JJ1[:,1])
plt.figure()
plt.plot(V_JJ1[:,0], V_JJ1[:,1])
```
